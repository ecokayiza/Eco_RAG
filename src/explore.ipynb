```python
from utils import *
from config import Config
import os

data_path = Config.DATA_DIR
example_path  = data_path / "C1" / "markdown"
all_files = [os.path.join(example_path, f) for f in os.listdir(example_path)]

def store(file,debug=True):
    try:
        _,ext = os.path.splitext(file)
        print(f"--------------------------------\nProcessing file: {file} with extension: {ext}")
        data = DataLoaderFactory.get_loader(ext).load_data(file)
        print(f"Loaded {file} with {len(data)} documents.")
        chunks = ChunkerFactory.get_chunker(ext).chunk(data)
        print(f"Chunked into {len(chunks)} chunks.")
        embeddings = HuggingFaceEmbedder.embed([chunk.page_content for chunk in chunks])
        print(f"Generated {len(embeddings)} embeddings.")
        VectorStore.store(embeddings)
        
    except Exception as e:
        if debug:
            print(f"Error processing {file}: {e}")

for file in all_files:
    store(file)
```
